```python
import pandas as pd
import geopandas as gpd
import seaborn as sns
import xarray as xr
import folium
```

```python
#importing the csv data from opendatanepal
Climate_data=pd.read_csv("climate_data_nepal_district_wise_monthly.csv")
Climate_data.head(5)
```

**Description of the Climate_data**

```python
print(Climate_data.describe())
```

**Cleaning the data**

```python
# Checking for any missing data
Climate_data.isnull().sum()
#renaming the columns 
Climate_data.rename(columns={
    "T2M": "Temperature",
    "PRECTOT": "Precipitation",
    "RH2M": "Humidity",
    "WS10M": "Wind_Speed_10m"
}, inplace=True)
print(Climate_data)
```

## <span style="color: #4421af;">Data Visualization</span>

 **<span style="color: darkteal;">Temperature Trend Map:</span>**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
Climate_data.groupby("YEAR")["Temperature"].mean().plot()
plt.xlabel("YEAR")
plt.ylabel("Average Temperature (Â°C)")
plt.title("Average Temperature Trends in Nepal (1981-2019)")
plt.show()
```

**Average Prectipitations trends in Nepal**

```python
plt.figure(figsize=(10, 5))
Climate_data.groupby("YEAR")["Precipitation"].mean().plot()
plt.xlabel("YEAR")
plt.ylabel("Precipitaion")
plt.title("Average Precipitation Trends in Nepal (1981-2019)")
plt.show()
```

**Scatter plot of temperature and prectipitations**

```python
sampled_data = Climate_data.sample(frac=0.3)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature', y='Precipitation', data=sampled_data ,s=30,alpha=0.5)
plt.title('Temperature vs. Precipitation')
plt.xlabel('Temperature')
plt.ylabel('Precipitation')
plt.show()
```

**Boxplot Temperature Yearly:**

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x=Climate_data["YEAR"], y=Climate_data["Temperature"], palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Temperature Distribution by Year")
plt.xlabel("Year")
plt.ylabel("Temperature (Â°C)")
plt.show()
```

**Heat map to find the pattern in Temperature**

```python
df_pivot = Climate_data.pivot_table(index="YEAR", columns="MONTH", values="Temperature", aggfunc="mean")

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_pivot, cmap="coolwarm", annot=False)
plt.title("Temperature Trend by Month (Â°C)")
plt.xlabel("Month")
plt.ylabel("Year")
plt.show()
```

## <span style="color: #7c1158;">Exploratory Data Analysis</span>

```python
# Compute basic statistics for relevant climate variables
climate_stats = Climate_data[['Temperature', 'Precipitation', 'Humidity', 'Wind_Speed_10m']].describe()

# For more specific statistics (e.g., mean, median)
mean_values = Climate_data[['Temperature', 'Precipitation', 'Humidity', 'Wind_Speed_10m']].mean()
median_values = Climate_data[['Temperature', 'Precipitation', 'Humidity', 'Wind_Speed_10m']].median()
min_values = Climate_data[['Temperature', 'Precipitation', 'Humidity', 'Wind_Speed_10m']].min()
max_values = Climate_data[['Temperature', 'Precipitation', 'Humidity', 'Wind_Speed_10m']].max()

# Print the basic statistics
print("Basic Statistics:\n", climate_stats)
print("\nMean Values:\n", mean_values)
print("\nMedian Values:\n", median_values)
print("\nMin Values:\n", min_values)
print("\nMax Values:\n", max_values)
```

**Data loading:**

```python
gdf = gpd.read_file('local_unit.shp')
print(gdf.head())
```

```python
from shapely.geometry import Point


geometry = [Point(xy) for xy in zip(Climate_data['LON'], Climate_data['LAT'])]
climate_gdf = gpd.GeoDataFrame(Climate_data, geometry=geometry, crs='EPSG:4326')
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, color='lightblue', edgecolor='black')
plt.title('All Districts')
plt.show()
```

```python
# Ensure both GeoDataFrames are in EPSG:4326
gdf = gdf.to_crs('EPSG:4326')
climate_gdf = climate_gdf.to_crs('EPSG:4326')
```

**Spatial Join:**

```python
# Spatial join between climate data and district polygons
merged_data = gpd.sjoin(climate_gdf, gdf, how='left', predicate='within')

# Check the first few rows to verify the merge
print(merged_data.head())
```

  **Data Cleaning:**

```python
# Group by district and calculate average precipitation
district_avg_precip = merged_data.groupby('DISTRICT_right')['Precipitation'].mean().reset_index()
```

```python
# Merge the precipitation data into the district GeoDataFrame
gdf_precip = gdf.merge(district_avg_precip, left_on='DISTRICT', right_on='DISTRICT_right', how='left')

# Optionally, remove rows with missing district values
gdf = gdf[gdf['DISTRICT'].notnull()]
district_avg_precip = district_avg_precip[district_avg_precip['DISTRICT_right'].notnull()]
```

```python
# Check for missing precipitation data
missing_precip = gdf_precip[gdf_precip['Precipitation'].isna()]
print(f"Missing precipitation data for {len(missing_precip)} regions")
```

```python
# Find districts with missing precipitation data
missing_precip = gdf_precip[gdf_precip['Precipitation'].isna()]
missing_districts = missing_precip['DISTRICT'].unique()
print(f"Missing precipitation data for the following districts: {missing_districts}")
```

```python
missing_districts = ['jhapa', 'bhojpur', 'khotang', 'siraha', 'ramechhap', 'sindhupalchok',
                     'dhading', 'bara', 'chitawan', 'baglung', 'nawalparasi_w', 'kapilbastu',
                     'pyuthan', 'rolpa', 'jajarkot', 'kalikot', 'bajura', 'achham', 'rukum_e']

missing_climate_data = climate_gdf[climate_gdf['DISTRICT'].isin(missing_districts)]
print(missing_climate_data)
```

```python
# Plot both district boundaries and climate data points to visually inspect
fig, ax = plt.subplots(figsize=(12, 10))
gdf.plot(ax=ax, color='lightgrey', edgecolor='black', alpha=0.5)  # Plot district boundaries
climate_gdf.plot(ax=ax, color='red', markersize=1)  # Plot climate data points
plt.title("District Boundaries and Climate Data Points")
plt.show()
```

**Choropleth Mapping:**

```python
# Plot average precipitation
fig, ax = plt.subplots(1, 1, figsize=(12, 10))
gdf_precip.plot(column='Precipitation', ax=ax, legend=True, cmap='BuPu',
                legend_kwds={'label': "Average Precipitation (mm)",
                                'orientation': "horizontal"})
ax.set_title('Average Precipitation by District')
plt.show()
```

```python
# Example plotting temperature.
district_avg_temp = merged_data.groupby('DISTRICT_right')['Temperature'].mean().reset_index()
gdf_temp = gdf.merge(district_avg_temp, left_on='DISTRICT', right_on='DISTRICT_right')

fig, ax = plt.subplots(1, 1, figsize=(12,10))
gdf_temp.plot(column = 'Temperature', ax=ax, legend=True, cmap='Reds', legend_kwds={'label': "Average Temperature (Â°C)", 'orientation': "horizontal"})
ax.set_title('Average Temperature by District')
plt.show()
```
